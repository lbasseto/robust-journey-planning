```python
import pandas as pd
from geopy import distance
```

```python
stations = pd.read_csv('data/bfkoordgeo.csv')
```

```python
# Filter stations based on distance from a coordinate (Zurich central by default)
def radius_filter_stations(radius_km, center_coord = (47.376846, 8.543938)):
    indexes_to_keep = []
    for idx, row in stations.iterrows():
        station_coord = (row['Latitude'], row['Longitude'])
        if distance.distance(center_coord, station_coord).km < radius_km:
            indexes_to_keep += [idx]

    return stations.iloc[indexes_to_keep]
```
